#Activation Functions effect on imgages


##Import Libraries


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from skimage import color
from skimage import io
```

##Tanch activation function

```python
def my_tanh(x):
  f_new=2/(1+np.exp(-2*x))
  f=f_new-1
  return f
```

```python
x=np.arange(-10,10,0.1)
y=my_tanh(x)
y_div=1-y**2
X_norm=x/10
plt.plot(x,y)
plt.plot(x,y_div)
#lt.plot(x,X_norm)
plt.grid()
```

##Softsign activation function

```python
def softsign(x):
  f=(x/(1+abs(x)))
  return f
```

```python
x=np.arange(-10,10,0.1)
y=softsign(x)
y_derv=(1/(1+np.abs(x))**2)
#x_norm=x/10
plt.plot(x,y)
plt.plot(x,y_derv)
#plt.plot(x,x_norm)
plt.grid()
```

##Sigmoid Activation

```python
def sigmoid(x):
  f=1/(1+np.exp(-x))
  return f
```

```python
x=np.arange(-10,10,0.1)
y=sigmoid(x)
y_derv=y*(1-y)
#x_norm=x/10
plt.plot(x,y)
plt.plot(x,y_derv)
#plt.plot(x,x_norm)
plt.grid()
```
